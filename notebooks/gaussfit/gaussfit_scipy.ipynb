```python
%matplotlib inline
import scipy as sp
import matplotlib.pyplot as plt
```

```python
def gauss(x, a, b, c):
    return a * sp.exp(-1*(x - b)**2/(2*c)**2)
```

```python
x = sp.arange(0,100,0.1)
y = gauss(x, 10, 50, 4) + sp.randn(len(x))*0.7
```

```python
plt.plot(x,y)
```

```python
from scipy.optimize import curve_fit
```

```python
curve_fit?
```

```python
pars, pcov = curve_fit(gauss, x , y, p0=[1, 1, 1])
```

```python
pars, pcov
```

```python
perr = sp.diag(pcov)**.5
perr
```

```python
plt.plot(x,y, label='data')
plt.plot(x,gauss(x,*pars))
None
```
