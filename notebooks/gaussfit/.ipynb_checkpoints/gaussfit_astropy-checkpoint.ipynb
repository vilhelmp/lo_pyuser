```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from astropy.modeling import models, fitting
```

```python
# Generate fake data
np.random.seed(0)
x = np.linspace(-5., 15., 4000)
y = 3 * np.exp(-0.5 * (x - 1.3)**2 / 0.8**2) + 6 * np.exp(-3 * (x - 10.3)**2 / 2**2)
y += np.random.normal(0., 0.2, x.shape)

# Fit the data using a box model
#t_init = models.Trapezoid1D(amplitude=1., x_0=0., width=1., slope=0.5)
#fit_t = fitting.LevMarLSQFitter()
#t = fit_t(t_init, x, y)

# Fit the data using a Gaussian
g_init = models.Gaussian1D(amplitude=1., mean=0, stddev=1.)
fit_g = fitting.LevMarLSQFitter()
g = fit_g(g_init, x, y)
```

```python
# Plot the data with the best-fit model
plt.figure(figsize=(8,5))
plt.plot(x, y, 'ko')
#plt.plot(x, t(x), 'b-', lw=2, label='Trapezoid')
plt.plot(x, g(x), 'r-', lw=2, label='Gaussian'.format())
plt.xlabel('Position')
plt.ylabel('Flux')
plt.legend(loc=2)
None
```

```python
%matplotlib inline
import scipy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gauss(x, a, b, c):
    return a * sp.exp(-1*(x - b)**2/(2*c)**2)
def gaussians(x, par):
    a1, a2, b1, b2, c1, c2 = par
    arr = gauss(x, a1, b1, c1) + gauss(x, a2, b2, c2)
    return arr
```

```python
x = np.arange(0,100,0.1)
#y = gauss(x, 10, 40, 4) + gauss(x, 50, 70, 6) + sp.randn(len(x))*0.7
# data
real_pars = [10, 50, 50, 70, 4, 6]
y = gaussians(x, real_pars) + sp.randn(len(x))*0.7
plt.plot(x,y)
None
```
