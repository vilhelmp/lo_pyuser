```python
%matplotlib inline
import scipy as sp
import numpy as np
import matplotlib.pyplot as plt
import pymc
import pymc.Matplot
```

```python
def gauss(x, a, b, c):
    return a * sp.exp(-1*(x - b)**2/(2*c)**2)
def gaussians(x, par):
    a1, a2, b1, b2, c1, c2 = par
    arr = gauss(x, a1, b1, c1) + gauss(x, a2, b2, c2)
    return arr
```

```python
x = np.arange(0,100,0.1)
#y = gauss(x, 10, 40, 4) + gauss(x, 50, 70, 6) + sp.randn(len(x))*0.7
# data
real_pars = [10, 50, 40, 70, 4, 6]
y = gaussians(x, real_pars) + sp.randn(len(x))*0.7
y_err = np.ones_like(y) * sp.randn(len(x))*0.7
plt.errorbar(x,y, yerr=y_err, ls='None', alpha=0.5)
None
```

```python
def make_model(x,y, y_err):
    #bnd = [(5,20), (40,60), (40,60), (60,80), (3,10), (3,10)]
    a1 = pymc.Uniform('a1', 5, 20, value=10)
    a2 = pymc.Uniform('a2', 30, 60, value=50)
    b1 = pymc.Uniform('b1', 30, 60, value=50)
    b2 = pymc.Uniform('b2', 60, 80, value=70)
    c1 = pymc.Uniform('c1', 3., 10, value=5)
    c2 = pymc.Uniform('c2', 3., 10, value=5)
    
    # model
    @pymc.deterministic(plot=False)
    def gauss(x=x, a1=a1, b1=b1, c1=c1, a2=a2, b2=b2, c2=c2):
        gone = a1 * sp.exp(-1*(x - b1)**2/(2*c1)**2) 
        gtwo = a2 * sp.exp(-1*(x - b2)**2/(2*c2)**2)
        return gone + gtwo
    # likelihood
    #~ y = pymc.Normal('y', mu=gauss, tau=1.0/amperr**2, value=f, observed=True)
    y = pymc.Normal('y', mu=gauss, tau=1/y_err**2, value=y, observed=True)
    return locals()
```

```python
M = pymc.MCMC(make_model(x,y,y_err))
```
