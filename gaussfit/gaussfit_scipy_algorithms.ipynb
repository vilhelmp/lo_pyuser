```python
%matplotlib inline
import scipy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gauss(x, a, b, c):
    return a * sp.exp(-1*(x - b)**2/(2*c)**2)
def gaussians(x, par):
    a1, a2, b1, b2, c1, c2 = par
    arr = gauss(x, a1, b1, c1) + gauss(x, a2, b2, c2)
    return arr
```

```python
x = np.arange(0,100,0.1)
#y = gauss(x, 10, 40, 4) + gauss(x, 50, 70, 6) + sp.randn(len(x))*0.7
# data
real_pars = [10, 50, 50, 70, 4, 6]
y = gaussians(x, real_pars) + sp.randn(len(x))*0.7
plt.plot(x,y)
None
```

```python
from scipy import optimize
def fitfunc(p, x, y):
    arr = np.nansum( (y - gaussians(x, p))**2 )
    return arr
p0=[10, 10, 50, 70, 4, 3]
bnd = [(5,20), (40,60), (40,60), (60,80), (3,10), (3,10)]
minimizer_kwargs = dict(method="L-BFGS-B",  bounds=bnd, args = (x, y))
result = optimize.minimize(fitfunc, p0, **minimizer_kwargs)
```
