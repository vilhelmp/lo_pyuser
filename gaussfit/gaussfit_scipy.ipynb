```python
%matplotlib inline
import scipy as sp
import matplotlib.pyplot as plt
```

```python
def gauss(x, a, b, c):
    return a * sp.exp(-1*(x - b)**2/(2*c)**2)
```

```python
x = sp.arange(0,100,0.1)
y = gauss(x, 10, 50, 4) + sp.randn(len(x))*0.7
```

```python
plt.plot(x,y)
```

```python
from scipy.optimize import curve_fit
```

```python
curve_fit?
```

```python
pars, pcov = curve_fit(gauss, x , y, p0=[1, 1, 1])
```

```python
pars, pcov
```

```python
perr = sp.diag(pcov)**.5
perr
```

```python
plt.plot(x,y, label='data')
plt.plot(x,gauss(x,*pars))
None
```

It fitted the local minima! 
So we have to specify some initial guesses.

```python
pars, pcov = curve_fit(gauss, x , y, p0=[5, 50, 3])
perr=sp.diag(pcov)**.5
plt.plot(x,y, label='data')
plt.plot(x,gauss(x,*pars), lw=4)
#plt.fill_between(x, gauss(x,))
None
```

```python
x = sp.arange(0,100,0.1)
y = gauss(x, 10, 40, 4) + gauss(x, 50, 70, 6) + sp.randn(len(x))*0.7
plt.plot(x,y)
None
```

```python
def gaussians(x, a1, a2, b1, b2, c1, c2):
    arr = gauss(x, a1, b1, c1) + gauss(x, a2, b2, c2)
    return arr
```

```python
curve_fit?
```

```python
pars, pcov = curve_fit(gaussians, x , y, p0=[10, 10, 50, 70, 4, 3])
perr = sp.diag(pcov)**.5
print pars, perr
plt.plot(x,y, label='data')
plt.plot(x, gauss(x, *pars[::2]))
plt.plot(x, gauss(x, *pars[1::2]))
plt.plot(x, gaussians(x,*pars), lw=4)
#plt.fill_between(x, gauss(x,))
None
```
